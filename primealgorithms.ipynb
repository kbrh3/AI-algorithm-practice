```python
def sieve_of_eratosthenesList(n):

  primes = [True] * (n + 1)  # assume all prime
  non_primes = [False] * (n + 1)  #assumed non-prime (except 0 and 1)
  non_primes[0] = non_primes[1] = True  # 0 and 1 are not prime
  p = 2

  while (p * p <= n):
    if primes[p]:
      # Marking multiples of p as non-prime
      for i in range(p * p, n + 1, p):
        primes[i] = False
        non_primes[i] = True
    p += 1

  # Extract prime nums from list
  primes = [i for i in range(2, n + 1) if primes[i]]

  return primes, non_primes

# Find primes up to given n
primes, non_primes = sieve_of_eratosthenesList(3000)

print("Prime numbers less than or equal to 100:")
print(primes)

#FIND prime number gaps
def find_gaps(data_list):
  differences = []
  for i in range(1, len(data_list)):
    difference = data_list[i] - data_list[i-1]
    if difference > 13:
      differences.append(difference)

  return differences

# Example usage
my_list = primes
difference_list = find_gaps(my_list)

print(f"List of differences: {difference_list}")

```

```python
n=100
def dijkstra_sieve(n):
    if n < 2:
        return []

    # Initialize the sieve with only odd numbers starting from 3
    sieve = [True] * ((n + 1) // 2)
    primes = [2]  # Start with 2 as the first prime

    for num in range(3, n + 1, 2):  # Consider only odd numbers
        if sieve[num // 2]:
            primes.append(num)
            for multiple in range(num * num, n + 1, 2 * num):
                sieve[multiple // 2] = False

    return primes

#n = 100
#primes = dijkstra_sieve(100)
#print(primes)
test_range = range(100, 1001, 100)

# Measure execution times
execution_times = [measure_time(dijkstra_sieve, n) for n in test_range]

plt.figure(figsize=(10, 6))
plt.plot(test_range, execution_times, marker='o', linestyle='-')
plt.title("Execution Time of dijkstra's sieve")
plt.xlabel("Number")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.show()

```

```python
def trial_division(limit):
    primes = []
    for num in range(2, limit+1):
        is_prime = True
        for prime in primes:
            if prime > int(num**0.5):
                break
            if num % prime == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)

    space = sys.getsizeof(primes)
    print(f'Space: {space/1e6} MB')
    return primes

def sieve_of_eratos(n):
    primes = []
    sieve = [True] * (n + 1)
    for x in range(2, int(n**0.5) + 1):
        if sieve[x]:
            for y in range(x*x, n + 1, x):
                sieve[y] = False
    for x in range(2, n + 1):
        if sieve[x]:
            primes.append(x)
    space = sys.getsizeof(primes) + sys.getsizeof(sieve)
    print(f'Space: {space/1e6} MB')
    return primes

def dijkstra(n):
    primes_pool = [(4, 2)]
    heapq.heapify(primes_pool)
    primes = [2]
    for i in range(3, n+1):
        while primes_pool[0][0] < i:
            multiple, prime = heapq.heappop(primes_pool)
            heapq.heappush(primes_pool, (multiple + prime, prime))
        if primes_pool[0][0] == i:
            multiple, prime = heapq.heappop(primes_pool)
            heapq.heappush(primes_pool, (multiple + prime, prime))
        else:
            primes.append(i)
            heapq.heappush(primes_pool, (i*i, i))
    space = sys.getsizeof(primes) + sys.getsizeof(primes_pool)
    print(f'Space: {space/1e6} MB')
    return primes

# Example usage
limit = 1_000_000

print('---Trial Division---')
t1 = time.perf_counter()
trial_division(limit)
print(f'Time: {time.perf_counter()-t1} seconds\n')

print('---Sieve of Eratosthenes---')
t1 = time.perf_counter()
sieve_of_eratos(limit)
print(f'Time: {time.perf_counter()-t1} seconds\n')

print('---Dijkstra\'s Approach---')
t1 = time.perf_counter()
dijkstra(limit)
print(f'Time: {time.perf_counter()-t1} seconds\n')
```

```python
import sys
import time
import heapq
import matplotlib.pyplot as plt

def trial_division(limit):
    primes = []
    for num in range(2, limit+1):
        is_prime = True
        for prime in primes:
            if prime > int(num**0.5):
                break
            if num % prime == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    space = sys.getsizeof(primes)
    return primes, space

def sieve_of_eratos(n):
    primes = []
    sieve = [True] * (n + 1)
    for x in range(2, int(n**0.5) + 1):
        if sieve[x]:
            for y in range(x*x, n + 1, x):
                sieve[y] = False
    for x in range(2, n + 1):
        if sieve[x]:
            primes.append(x)
    space = sys.getsizeof(primes) + sys.getsizeof(sieve)
    return primes, space

def dijkstra(n):
    primes_pool = [(4, 2)]
    heapq.heapify(primes_pool)
    primes = [2]
    for i in range(3, n+1):
        while primes_pool[0][0] < i:
            multiple, prime = heapq.heappop(primes_pool)
            heapq.heappush(primes_pool, (multiple + prime, prime))
        if primes_pool[0][0] == i:
            multiple, prime = heapq.heappop(primes_pool)
            heapq.heappush(primes_pool, (multiple + prime, prime))
        else:
            primes.append(i)
            heapq.heappush(primes_pool, (i*i, i))
    space = sys.getsizeof(primes) + sys.getsizeof(primes_pool)
    return primes, space

# Example usage
limits = [10_000, 100_000, 500_000, 1_000_000]
methods = [trial_division, sieve_of_eratos, dijkstra]
times = {method.__name__: [] for method in methods}
spaces = {method.__name__: [] for method in methods}

for limit in limits:
    for method in methods:
        t1 = time.perf_counter()
        _, space = method(limit)
        t2 = time.perf_counter()
        times[method.__name__].append(t2 - t1)
        spaces[method.__name__].append(space)

# Plotting Time Complexity
plt.figure(figsize=(12, 6))
for method in methods:
    plt.plot(limits, times[method.__name__], label=method.__name__, marker='o')
plt.title("Time Complexity of Prime Generation Methods")
plt.xlabel("Upper Limit")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()

# Plotting Space Complexity
plt.figure(figsize=(12, 6))
for method in methods:
    plt.plot(limits, [space / 1e6 for space in spaces[method.__name__]], label=method.__name__, marker='o')
plt.title("Space Complexity of Prime Generation Methods")
plt.xlabel("Upper Limit")
plt.ylabel("Space (MB)")
plt.legend()
plt.grid(True)
plt.show()
```

```python

def miller_rabin(n, k=5):
    """Use the Miller-Rabin primality test to determine if n is prime."""
    # Handle base cases
    if n == 2 or n == 3:
        return True
    if n <= 1 or n % 2 == 0:
        return False

    # Write n as 2^r * d + 1 with d odd (by factoring out powers of 2 from n - 1)
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    # Witness loop
    for _ in range(k):
        a = random.randint(2, n - 2)
        x = pow(a, d, n)  # Compute a^d % n using efficient modular exponentiation
        if x == 1 or x == n - 1:
            continue

        for _ in range(r - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False  # n is definitely composite

    return True  # n is probably prime

# Test the function
n = 31
print(f"Is {n} a prime number? {miller_rabin(n)}")
n = 32
print(f"Is {n} a prime number? {miller_rabin(n)}")

def measure_time(func, n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

# Generate a range of numbers to test
test_range = range(100, 1001, 100)
#testing the range 100,200,300,400,500,600,700,800,900,1000

# Measure execution times
execution_times = [measure_time(miller_rabin, n) for n in test_range]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(test_range, execution_times, marker='o', linestyle='-')
plt.title("Execution Time of Miller Rabin Algorithm")
plt.xlabel("Number")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import time
import random
def jacobi_symbol(a, n):
    """Compute the Jacobi symbol (a/n) using the algorithm."""
    assert n > 0 and n % 2 == 1, "n must be a positive odd number."
    a = a % n
    jacobi = 1
    while a != 0:
        while a % 2 == 0:
            a //= 2
            if n % 8 in [3, 5]:
                jacobi = -jacobi
        a, n = n, a  # Swap a and n
        if a % 4 == 3 and n % 4 == 3:
            jacobi = -jacobi
        a %= n
    return jacobi if n == 1 else 0

def solovay_strassen(n, k=10):
    """Check if n is a prime number using the Solovay-Strassen primality test."""
    if n == 2:
        return True
    if n < 2 or n % 2 == 0:
        return False
    for _ in range(k):
        a = random.randrange(2, n - 1)
        x = jacobi_symbol(a, n)
        if x == 0 or pow(a, (n - 1) // 2, n) != x % n:
            return False
    return True

# Test the function
n = 31
print(f"Is {n} a prime number? {solovay_strassen(n)}")
n = 32
print(f"Is {n} a prime number? {solovay_strassen(n)}")

def measure_time(func, n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

# Generate a range of numbers to test
test_range = range(100, 1001, 100)
#testing the range 100,200,300,400,500,600,700,800,900,1000

# Measure execution times
execution_times = [measure_time(solovay_strassen, n) for n in test_range]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(test_range, execution_times, marker='o', linestyle='-')
plt.title("Execution Time of Solovay-Strassen Primality Test")
plt.xlabel("Number")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.show()
```
